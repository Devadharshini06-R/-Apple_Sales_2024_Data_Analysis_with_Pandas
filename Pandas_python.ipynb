# Data cleaning

```python
df.duplicated().sum() #count of duplicate rows
```

```python
df.fillna(0,inplace=True) # fill all missing values to 0
df
```

```python
df.replace({'Europe':'EU','Greater China':'China'},inplace=True) #  Change the name in particular column
df
```

```python
# rename column to make easier to use
df.rename(columns={'iPhone Sales (in million units)':'iphone_sales',
                   'iPad Sales (in million units)':'ipad_sales',
                   'Mac Sales (in million units)':'mac_sales',
                   'Wearables (in million units)':'wearables',
                   'Services Revenue (in billion $)':'Service_revenue'},inplace=True)
df.head()
```

```python
# convert columns to float types for numeric operations
df=df.astype({'iphone_sales':'float',
              'ipad_sales':'float',
              'mac_sales':'float',
              'wearables':'float',
              'Service_revenue':'float'})
df
```

# Aggregation & Grouping

## Groupby one column

```python
df.groupby('Region').size()
```

```python
# total sum of all numeric columns grouped by state
df.groupby('State').sum()
```

```python
# groupby both region and state then get sum
df.groupby('Region')['State'].sum()
```

## range

```python
# range (max-min) of iPad sales for each region
df.groupby('Region')['ipad_sales'].max() - df.groupby('Region')['ipad_sales'].min()
```

## Groupby by multiple columns

```python
#  groupby both region and state, then get average
df.groupby(['Region','State']).mean()
```

## custom agregations with agg()

```python
# sum and average
df.groupby('Region').agg({'iphone_sales':'sum','ipad_sales':'mean'})
```

```python
# get min,max and mean for all numeric columns per region
df.groupby('Region')['ipad_sales'].agg(['min','max','mean'])
```

## value counts ( like group counts)

```python
# count how many times each region appears
df['Region'].value_counts()
```

```python
# same as value_counts but works with groupby
df.groupby('State').value_counts()
```

## using pivot_table()

```python
# create pivot table : region as rows state as columns
df.pivot_table(values='iphone_sales',index='Region',columns='State')
```

```python
# pivot with total iphone sales per state
df.pivot_table(values='iphone_sales',index='State',aggfunc='sum')
```

## crosstab for frequency

```python
# frequency table: how many times each region-state pair occurs
pd.crosstab(df['Region'],df['State'])
```

# Merging and joining

```python
df1=pd.read_csv('/content/samsung_sales_1000_rows.csv')
df1
```

## Concatenation

```python
# stack two dataframe vertically ( same columns)
pd.concat([df1,df])
```

```python
# stack tow dataframe side by side (column wise)
pd.concat([df1,df],axis=1)
```

## Merge by key column

```python
# merge two dataframe where state column matches
pd.merge(df,df,on='State')
```

```python
# left join (keep all rows from first df)
pd.merge(df,df,how='left',on='State')
```

```python
# right join (keep all the rows from second df)
pd.merge(df,df,how='right',on='State')
```

```python
# outer join (keep all the rows from both df)
pd.merge(df,df,how='outer',on='State')
```

```python
# inner join (only matching rows)
pd.merge(df,df,how='inner',on='State')
```

## Merge with different column names

```python
#  merge where df['sate'] == df['state']
pd.merge(df,df,left_on='State',right_on='State')
```

## Indicator for join type

```python
# add a column showing the source of each row: left_only, right_only,both
pd.merge(df,df,on='State',how='outer',indicator=True)
```
